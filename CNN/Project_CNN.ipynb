```python
import glob
import numpy as np
import pandas as pd
from PIL import Image
from numpy import asarray

# Declare an empty list
testing_array = []

# Using a for loop to append each image to the list
for filename in glob.glob('/home/uca/Desktop/Undergrad/Arfan_FinalYearProject/CNN/test/*.*'):
    im = Image.open(filename)
    array = asarray(im)
    image_data = array.flatten()
    testing_array.append(image_data)

# Convert the list to a numpy array
testing_array = np.array(testing_array)

# Convert the numpy array to a pandas DataFrame
dataset = pd.DataFrame(testing_array)

# Save the DataFrame as a CSV file
dataset.to_csv('Testing_images.csv', sep=',', index=False, header=False)
```

```python
#datasets
x_train = np.loadtxt('Training_images.csv', delimiter = ',')
y_train = np.loadtxt('Training_Labels.csv', delimiter = ',')

x_test = np.loadtxt('Testing_images.csv', delimiter = ',')
y_test = np.loadtxt('Testing_Labels.csv', delimiter = ',')

print("shape of x_train is ", x_train.shape)
print("shape of y_train is ", y_train.shape)
print("shape of x_test is ", x_test.shape)
print("shape of y_test is ", y_test.shape)
```

```python
#lets reshape them
x_train = x_train.reshape(len(x_train), 256, 256, 3)
y_train = y_train.reshape(len(y_train), 1)
x_test = x_test.reshape(len(x_test), 256, 256, 3)
y_test = y_test.reshape(len(y_test), 1)

print("shape of x_train is ", x_train.shape)
print("shape of y_train is ", y_train.shape)
print("shape of x_test is ", x_test.shape)
print("shape of y_test is ", y_test.shape)
```

```python
x_train[1, :]
```

```python
#rescale to train model properly
x_train = x_train/255.0
y_train = y_train/255.0
x_test = x_test/255.0
y_test = y_test/255.0
```

```python
x_train[1, :]
```

```python
import matplotlib.pyplot as plt
import random
img1 = random.randint(0, len(x_train))
plt.imshow(x_train[img1, :])
plt.show()
```

```python
#import required libraries to build the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten

model = Sequential([
    Conv2D(32, (3, 3), activation = 'relu', input_shape = (256, 256, 3)),
    MaxPooling2D((2, 2)),
    
    Conv2D(32, (3, 3), activation = 'relu'),
    MaxPooling2D((2, 2)),
    
    Flatten(),
    Dense(64, activation = 'relu'),
    Dense(1, activation = 'sigmoid')

])
```

```python
#compile the model
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 20, batch_size = 64, verbose = 1, validation_data = (x_test, y_test))
```

```python
#save the model
model.save('imageClassificationCNN.h5')
```

```python
from tensorflow.keras.models import load_model

lm = load_model('imageClassificationCNN.h5')
lm.summary()
```

```python
model.evaluate(x_test, y_test)
```

```python
#make predicions
img2 = random.randint(0, len(x_test))
plt.imshow(x_test[img2, :])
plt.show()

y_pred = model.predict(x_test[img2, :].reshape(1, 256, 256, 3))


if (y_pred <= 0.5):
    pred = 'healthy'
else:
    pred = 'unhealthy'
    

print("Model says it is ", pred)
```
