```python
# Importing the Keras libraries and packages
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from PIL import Image

# Initialising the CNN
CNN_Classifier=Sequential();

# Step 1 - Convolution
CNN_Classifier.add(Conv2D(32,(3,3),input_shape=(64,64,3), activation='relu'))

# Step 2 - Pooling
CNN_Classifier.add(MaxPooling2D(pool_size=(2,2)))


# Step 3 - Flattening
CNN_Classifier.add(Flatten())


# Step 4 - Full connection
CNN_Classifier.add(Dense(128, activation='relu'))


# Adding the Output Layer
CNN_Classifier.add(Dense(10, activation='softmax'))

# Compiling the CNN
CNN_Classifier.compile(optimizer ='adam',loss='categorical_crossentropy',metrics=['accuracy'])

from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)

training_set = train_datagen.flow_from_directory('Data/Train',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')

test_set = test_datagen.flow_from_directory('Data/Test',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'categorical')

CNN_Classifier.fit_generator(training_set,
                         steps_per_epoch = 70,
                         epochs = 15,
                         validation_data = test_set,
                         validation_steps = 50)


# prediction of the image
from tensorflow.keras.preprocessing import image
test_image = image.load_img ('Data/Single/D.JPG', target_size= (64, 64))
test_image.show()
test_image = image.img_to_array (test_image)
test_image = np.expand_dims(test_image, axis = 0)
results = CNN_Classifier.predict(test_image)
training_set.class_indices
if results[0][0] == 1:
    prediction = 'A'
    print(prediction)
elif results[0][1] == 2:
    prediction = 'B'
    print(prediction)
elif results[0][2] == 3:
    prediction = 'C'
    print(prediction)
elif results[0][3] == 4:
    prediction = 'D'
    print(prediction)
elif results[0][4] == 5:
    prediction = 'E'
    print(prediction)
elif results[0][5] == 6:
    prediction = 'F'
    print(prediction)
elif results[0][6] == 7:
    prediction = 'G'
    print(prediction)
elif results[0][7] == 8:
    prediction = 'H'
    print(prediction)
elif results[0][8] == 9:
    prediction = 'I'
    print(prediction)
else:
    prediction = 'J'
    print(prediction)
    
```
