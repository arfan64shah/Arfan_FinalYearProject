```python
#import required libraries
import pandas as pd
import numpy as np

dataset = pd.read_csv('Output50.csv')
dataset2 = pd.read_csv('Labels50.csv')
dataset.head()
```

```python
#remove the first column
dataset1 = dataset.drop(['Unnamed: 0'], axis=1)
dataset1.head()
```

```python
#distribute data into dependent and independent variables of x and y
x = dataset1.iloc[:, :]
y = dataset2.iloc[:, :]

#now split data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.16, random_state = 0)

#apply the model and predict
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

model = AdaBoostClassifier(DecisionTreeClassifier(), n_estimators=100, learning_rate=1.0)

model.fit(x_train, y_train)

pred = model.predict(x_test)

#now find accuracy
from sklearn.metrics import accuracy_score

print("Accuracy is ", accuracy_score(y_test, pred))
```
