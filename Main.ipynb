```python
#now read the images one by one using for loop
nothealthy_image_list = []

#make a path
path = "C:/Users/arfan.shah/Desktop/UCA_Undergrad/Arfan_FinalYearProject/data/nothealthy/*.*"

#use a for loop to read images one by one
for image in glob.glob(path):
    
    #now read the image and store it as numpy array
    images = cv2.imread(image)
    nothealthy_image_list.append(images)
    
health_image_list
```

```python
#make an empty list of images
import pandas as pd
df = pd.DataFrame(columns=["Image"])

for img_name in nothealthy_image_list:
    img = cv2.imread(img_name)
    img_str = img.tostring()
    df = df.append({"Image": img_str}, ignore_index=True)

    
writer = pd.ExcelWriter("images.xlsx", engine="xlsxwriter")
df.to_excel(writer, sheet_name="Images", index=False)
writer.save()

```
