```python
#import the libraries
import numpy as np
import pandas as pd

#load the data
dataset = pd.read_csv('Final.csv')
dataset.head()
```

```python
#number of classes in the target variable
dataset['Label'].value_counts()
```

```python
#classify the dataset into dependent and independent sets
x = dataset.iloc[:, : -1]
y = dataset.iloc[:, -1]


print(x.shape)
```

```python
#now split data into training and testing
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)


#import the clssifier
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

#fit the model
gnb.fit(x_train, y_train)

#predict outcome
pred = gnb.predict(x_test)

#now lets measure the accuracy and confussion matrix
#lets now find the accuracy
from sklearn.metrics import accuracy_score
print("The accuracy for main data is ", accuracy_score(y_test, pred))

#confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, pred))

#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```
