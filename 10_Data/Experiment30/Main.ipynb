```python
#import the required libraries
import skimage.feature as skf
import openpyxl
import numpy as np
import math
import cv2
from PIL import Image
from numpy import asarray
from scipy.stats import skew, kurtosis
import skimage.feature
import xlsxwriter
import os


#creates a file for red channel
outWorkbookR = xlsxwriter.Workbook("Main30.xlsx")
outSheetR = outWorkbookR.add_worksheet()

#writes headers for red file
outSheetR.write("A1", "Mean")
outSheetR.write("B1", "RMS")
outSheetR.write("C1", "Median")
outSheetR.write("D1", "Variance")
outSheetR.write("E1", "Std")
outSheetR.write("F1", "Skewness")
outSheetR.write("G1", "Kurtosis")
outSheetR.write("H1", "Contrast")
outSheetR.write("I1", "Entropy")
outSheetR.write("J1", "Energy")
outSheetR.write("K1", "Homogeneity")
outSheetR.write("L1", "Correlation")
outSheetR.write("M1", "IDM")
outSheetR.write("N1", "Smoothness")
outWorkbookR.close()
```

```python
#import required libaray to read and make a list of all the needy images
import glob

#declare an empty list
Image_list = []

#using for loop to append to the list declared above
for filename in glob.glob('/home/uca/Desktop/Undergrad/Arfan_FinalYearProject/10_Data/Experiment30/experimental_30/*.*'):
    im = Image.open(filename)
    
    #append each image to the list declared above
    Image_list.append(im)
    
#prints number of images used
print(len(Image_list))
```

```python
#uses for loop to read each image and array of each image
for i in range(0, len(Image_list)):
    Image = Image_list[i]
    image = asarray(Image)
  
    
#     #Here i am extracting the single channel like red, green, and blue from the whole image array
#     R = image[:, :, 0]
#     G = image[:, :, 1]
#     B = image[:, :, 2]



    #selecting features from Red Channel
    print("Calculating Statistical features: ")
    
    #calculates mean
    Mean = np.mean(image);
    print(Mean)
    
    #calculating root mean square
    RMS = np.sqrt(np.mean(image**2))
    print(RMS)
    
    #calculating median
    Median = np.median(image)
    print(Median)
    
    #calculating variance of the pixels
    Variance = np.var(image)
    print(Variance)
    
    #calculating standard deviation
    Std = math.sqrt(Variance)
    print(Std)
    
    #calculates skewness
    Skewness = skew(image.reshape(-1))
    print(Skewness)
    
    #calculates kurtosis
    Kurtosis = kurtosis(image.reshape(-1))
    print(Kurtosis)
    
    #calculates entropy
    Entropy = skimage.measure.shannon_entropy(image)
    print(Entropy)
    
    #returns an image array where lowest pixel value is 0 and highest 8 instead of 0 and 255
    image = skimage.img_as_ubyte(image)
    
    #calculates contrast
    g1 = skf.graycomatrix(image, [1], [0], levels=256, symmetric=False, normed=True)
    Contrast= skf.graycoprops(g1, 'contrast')[0][0]
    print(Contrast)
    
    #calculates energy
    Energy = skf.graycoprops(g1, 'energy')[0][0]
    print(Energy)
    
    #calculating homogeniety
    Homogeneity = skf.graycoprops(g1, 'homogeneity')[0][0]
    print(Homogeneity)
    
    #finding correlation
    Correlation = skf.graycoprops(g1, 'correlation')[0][0]
    print(Correlation)
    
    #glcm_R = skf.graycomatrix(R, distances=[1], angles=[0], levels=256, symmetric= False, normed=True)
    
    #calculate inverse difference moment (IDM)
    IDM = 1 / (1 + Contrast - Energy)
    print(IDM)
    
    #calculates smoothness
    diff = np.diff(image.astype(float))
    variance = np.var(diff)
    Smoothness = variance / (image.shape[0] * image.shape[1])
    print(Smoothness)
    
    #count the number of images
    print("Image number: ", i)
    
    
    #Declares data for all the three channels
    values = [Mean, RMS, Median, Variance, Std, Skewness, Kurtosis, Entropy, Contrast, Energy, Homogeneity, Correlation, IDM, Smoothness]
    
    #creates a general fle where all the three channels data will be merged
    outWorkbook = xlsxwriter.Workbook("Main30.xlsx")
    outSheet = outWorkbook.add_worksheet()
    
    #writes into the general file
    outWorkbook = openpyxl.load_workbook("Main30.xlsx") 
    outSheet = outWorkbook.active
    outSheet.append(values)
    
    #here the general file is being saved
    outWorkbook.save(filename="Main30.xlsx")
```
